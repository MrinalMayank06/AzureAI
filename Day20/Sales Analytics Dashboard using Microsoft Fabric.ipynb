```python
from pyspark.sql.functions import when, col

sales_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/fabric_advanced_sales.csv")

sales_df = sales_df.withColumn(
    "Sales_Category",
    when(col("Sales") > 5000, "High").otherwise("Low")
)

sales_df.write.mode("overwrite").saveAsTable("sales_table")

customer_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/customer.csv")

customer_df.write.mode("overwrite").saveAsTable("customer_table")

product_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/product.csv")

product_df.write.mode("overwrite").saveAsTable("product_table")
```
