```sql
-- Welcome to your new notebook
-- Type here in the cell editor to add code!
patients_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/patients.csv")
```

```sql
doctors_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/doctors.csv")
```

```sql
appointments_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("Files/appointments.csv")
```

```sql
patients_df.write.mode("overwrite").saveAsTable("patient_table")
doctors_df.write.mode("overwrite").saveAsTable("doctors_table")
appointments_df.write.mode("overwrite").saveAsTable("appointments_table")
```
