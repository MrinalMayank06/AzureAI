# 🚑 Healthcare Data Surge Optimization
End-to-End PySpark Streaming Simulation

## Context
Real-time ICU patient monitoring with doubled data frequency (0.5 sec).

## Dataset Preview

```python
import pandas as pd
df = pd.read_csv("/content/healthcare_stream_dataset.csv")
df.head()
```

## Basic Statistics

```python
df.describe()
```

## 🚨 Alert Detection Logic

```python
alerts = df[(df['heart_rate'] > 120) | (df['oxygen_level'] < 90)]
alerts.head()
```

## 📊 Aggregation (Patient-wise Avg)

```python
agg = df.groupby('patient_id').agg({
    'heart_rate': 'mean',
    'oxygen_level': 'mean'
}).reset_index()

agg.head()
```

## 💾 Save Processed Data

```python
agg.to_csv('processed_output.csv', index=False)
print("Saved!")
```
